Data Anlysis and Visualization

```python
cancellation=hotel_data['is_canceled'].value_counts()
print(cancellation)

plt.figure(figsize=(7,4))
plt.title("Booking Cancellation Status")
plt.bar(['Not Cancelled','Cancelled'],hotel_data['is_canceled'].value_counts(),edgecolor='b',width=0.89,color='green')
plt.show()
```

Cancellation ans Non Cancellation ratios of City Hotels and Resort Hotels

```python
import seaborn as sns
plt.figure(figsize=(5,6))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=hotel_data,palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
plt.title("Status of City Hotels and Resort Hotels")
plt.xlabel('Hotels')
plt.ylabel('Number of Reservations')
```

Resort Hotel details

```python
resort_hotel=hotel_data[hotel_data['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

City Hotel details

```python
city_hotel=hotel_data[hotel_data['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

We need to check whether the price is effecting the hotels,so we need to check the mean by the two hotels with reservation_status_date

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(9,4))
plt.title("Avg Daily rate in City and Resort Hotels")
plt.plot(resort_hotel.index,resort_hotel['adr'])
plt.plot(city_hotel.index,city_hotel['adr'])
```

We need to see the more cancellation rates in months

```python
hotel_data['month']=hotel_data['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
sns.countplot(x='month',hue='is_canceled',data=hotel_data)
```

```python
plt.figure(figsize=(15,8))
plt.title("ARD")
d1=hotel_data[hotel_data['is_canceled']==1].groupby('month')[['adr']].sum().reset_index()
plt.plot(hotel_data['month'],hotel_data['adr'])
plt.show()
```

Cancellation rates in Countries

```python
cancelled_data=hotel_data[hotel_data['is_canceled']==1]
top_10_countries=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 Countries with High Cancellation rates')
plt.pie(top_10_countries,autopct='%.2f',labels=top_10_countries.index);
```

```python
cancelled_data_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_data_adr.reset_index(inplace=True)
cancelled_data_adr.sort_values('reservation_status_date',inplace=True)

Non_Cancelled_data=hotel_data[hotel_data['is_canceled']==0]
Non_Cancelled_data_adr=Non_Cancelled_data.groupby('reservation_status_date')[['adr']].mean()
Non_Cancelled_data_adr.reset_index(inplace=True)
Non_Cancelled_data_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title("Average Daily Rates")
plt.plot(Non_Cancelled_data_adr['reservation_status_date'],Non_Cancelled_data_adr['adr'],label="Non Cancelled",color="red")
plt.plot(cancelled_data_adr['reservation_status_date'],cancelled_data_adr['adr'],label="Cancelled",color="blue")
plt.legend()
```
